```python
#Création d'une fonction afin d'obtenir des information de base paour les jeux de données(taille, pourcentage données
# manquantes par colonne et nb lignes dupliquées)
def qualit(donnees):
    print("nb de lignes x nb de colonnes : ",donnees.shape,".\n")
    print("% de données manquantes :\n",(donnees.isna().sum()/len(donnees) * 100).sort_values(ascending = False),"\n")
    print("nb de lignes dupliquées :",donnees.duplicated().sum(),".\n")
```

### Jeu de données Barro-Lee
L'ensemble de données Barro-Lee donne une image complète de la répartition du niveau

```python
#application d'un filtre afin de ne conserver que les indicateurs utiles (liste 'col_util') de la colonne 'indicator Name'
data_util = data3[data3['Indicator Name'].isin(col_util)]
qualit(data_util)
```

```python
#retrait des colonnes vides
data_util = data_util[data_util.columns[data_util.isna().sum()/len(data_util) != 1]]
```

```python
#Groupes de pays faisont partie d'une 'Region' de data1
data1['Region'].unique()
```

```python
#Groupes ne faisont pas partie de 'Region' dans data1
#verification des nan, gibraltar et nauru n ont pas de code region, à verifier et eventuellement completer
data1[data1['Region'].isna()]
```

```python
#verification des information data1 nécessaires à la creation d'une colonne Regions par jointure avec "Country Code"
data1[['Region','Country Code']]
```

```python
#ajout de la colonne Region par jointure
data_util = pd.merge(data1[['Region','Country Code']], data_util, how="right", on=["Country Code"])
data_util
```

```python
#eliminations de colonnes non nécessaires
data_util = data_util.drop(columns=['Country Code','Indicator Code'])
```

```python
#creation de 'list_Country' qui contiendra les noms de pays et regions de la colonne 'Country Name'
#divisible en 2 parties (pays, regions)
list_Country = data_util['Country Name'].unique()
list_Country
```

```python
#regions de 'list_country'
list_Country[:25]
```

```python
#confirmation de la liste des regions obtenue sur data1, les regions de list country qui different ne representant 
#pas de regions geographiques ou qui ne sont pas pertinentes, nous ne les conserverons pas.
reg_list = ['East Asia & Pacific',
 'Europe & Central Asia',
 'Latin America & Caribbean',
 'Middle East & North Africa',
 'North America',
 'South Asia',
 'Sub-Saharan Africa']
```

```python
len(reg_list)#nombre de regions
```

```python
# Creation d'un dataset principal df
```

```python
#pivot et transposition pour obtenir un dataset df avec les variables aux emplacements voulus
df = data_util.pivot(index=['Indicator Name'],columns =['Region','Country Name'])
df
```

```python
df = df.T.reset_index()
```

```python
df.columns.name = None
df.rename(columns={"level_0":"Year"},inplace=True)
df
```

```python
df.Year.dtypes
```

```python
df.Year = df.Year.astype('int')
```

## creation de 'data_util_reg', qui ne contiendra que les donnees des regions de la colonne 'Country Name' a partir de la liste obtenue sur data1.

```python
data_util_reg = df[df['Country Name'].isin(reg_list)]
qualit(data_util_reg)
```

```python
data_util_reg.head(5)
```

```python
for column in data_util_reg.columns[3:]:
    print(print(data_util_reg[column].value_counts().sort_values(ascending = False)))
```

```python
#retrait des colonnes vides
data_util_reg = data_util_reg[data_util_reg.columns[data_util_reg.isna().sum()/len(data_util_reg) != 1]]
data_util_reg
```

```python
#Moyennes des regions de la colonne "Country Name"
data_util_reg.groupby(["Country Name"]).mean().iloc[:,1:]
```

```python
#Ecart-types des regions de la colonne "Country Name"
data_util_reg.groupby(["Country Name"]).std().iloc[:,1:]
```

### pourcentage d'information disponibe par variable et années (sur les groupes de pays deja present sur la colonne 'Country Name', 6 dernieres années )

```python
disp_reg = data_util_reg.groupby(["Year"]).count().tail(6)
disp_reg = disp_reg[disp_reg.columns[disp_reg.sum() != 0]]
disp_reg = round((disp_reg/7)*100,2)
disp_reg.iloc[:,1:]
```

```python
plt.style.use('default')
plt.rcParams["figure.figsize"]=[13,6]
sns.lineplot(data=data_util_reg, x="Year", y='Internet users (per 100 people)', hue="Country Name", markers=True, dashes=False, style="Country Name")
plt.xlabel("Année",fontsize=20)
plt.ylabel("% utilisateurs",fontsize=20)
plt.title("Evolution du pourcentage d'utilisateurs internet par région", fontsize=20)
plt.xticks(fontsize = 13, rotation = 45,c='k')
plt.yticks(fontsize = 13)
#plt.savefig("net_reg.jpg",bbox_inches ="tight")
plt.draw()
```
